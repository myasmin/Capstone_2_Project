### 1. Importing the Libraries:

Using Opendatasets Library:

```python
# pip install opendatasets

import opendatasets as od
import pandas as pd

kaggle_dataset="https://www.kaggle.com/datasets/shilongzhuang/telecom-customer-churn-by-maven-analytics"
od.download(kaggle_dataset)
```

### 2. Read the dataset:

```python
churndata = pd.read_csv('./telecom-customer-churn-by-maven-analytics/telecom_customer_churn.csv')
```

#### Find null/ missing values

```python
missing = pd.concat([churndata.isnull().sum(), 100 * churndata.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count',ascending=False)
```
